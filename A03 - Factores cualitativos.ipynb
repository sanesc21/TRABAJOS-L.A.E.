Ahora convertimos a 1 y 0

```python
import pandas as pd

df = pd.read_excel('Datos_Aguas.xlsx')


df['Tamaño'] = df['Tamaño'].map({'Grande': 1, 'Chica': 0})
df['Hielo'] = df['Hielo'].map({'Con': 1, 'Sin': 0})


df = pd.get_dummies(df, columns=['Sabor'], prefix='', prefix_sep='', dtype=int)

df = df[['Tamaño', 'Hielo', 'Kalua', 'Horchata', 'Score']]

print(df.head())
```

```python
from sklearn.linear_model import LinearRegression

X = df[['Tamaño', 'Hielo', 'Kalua', "Horchata"]]
y = df['Score']

modelo_lr = LinearRegression()
modelo_lr.fit(X, y)

print(f"Coeficientes: {modelo_lr.coef_}")
print(f"R^2: {modelo_lr.score(X, y):.4f}")
```
